```python
df['reviews'] = df['reviews'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['reviews'].head() #lowercase
```

```python
df['reviews'] = df['reviews'].str.replace('[^\w\s]','')
df['reviews'].head() #removing punctuation
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['reviews'] = df['reviews'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df['reviews'].head()
```

```python
freq = pd.Series(' '.join(df['reviews']).split()).value_counts()[:50]
freq #most words
```

```python
tf1 = (df['reviews'][1:2]).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf1.columns = ['words','tf']
tf1
#Term frequency specifies how frequently a term appears in the entire document.
#It can be thought of as the probability of finding a word within the document.
```

```python
import numpy as np
```

```python
for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(df.shape[0]/(len(df[df['reviews'].str.contains(word)])))


tf1
#The more the value of IDF, the more unique is the word.
```

```python
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```
