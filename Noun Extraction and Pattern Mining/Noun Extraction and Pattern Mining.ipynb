### Function to convert list to string

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1
```

### Performing this below looping code to exract nouns

```python
spell = SpellChecker()

tagged_list=[]
words_list_full=[]
#tagged_list=[]
m = 0
for j in range(len(review['reviews'])):
    paragraph = listToString(review['reviews'][j])
    para_noun = re.sub(r'[^\w\s]', '', paragraph.lower()) 
    
    #tokenized = sent_tokenize(txt)
    
    tokenized = sent_tokenize(para_noun) 

    #print(tokenized)
    
    
    for i in tokenized: 
        #global tagged
        
        words_list=[]
        # Word tokenizers is used to find the words  
        # and punctuation in a string 
        wordsList = nltk.word_tokenize(i) 
  
        # removing stop words from wordList 
        wordsList = [w for w in wordsList if not w in stop_words] 
        wordsList = [w for w in wordsList if w in english_words_set]  
        misspelled = spell.unknown(wordsList)
        #print(misspelled)
        
        wordsList = [w for w in wordsList if w not in misspelled]  
        wordsList = list(set(wordsList))
    
        words_list.append(wordsList)
        
        words_list_full = words_list_full + words_list
  
        #  Using a Tagger. Which is part-of-speech  
        # tagger or POS-tagger.
        
        tagged = nltk.pos_tag(wordsList)
        #print(tagged) 
        tagged_list.append(tagged)
```

```python
## There are chances of same nouns to be reused in the different reviews. Hence we need to pick all 
## nouns in a list and perform a set operation on them

words_list_full_tagged = []
for i in words_list_full: 
    tagged = nltk.pos_tag(i)
    #print(tagged)
    words_list_full_tagged.append(tagged)
    
    #tagged_list.append(tagged)
```
