```python
import numpy as np
import pandas as pd
df = pd.read_excel("/content/Reviews_information_cleaned_data_final.xlsx")
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
## The matcher module from spacy is used for pattern based matching. We define the pattern as per our requirement 
## and the matching is performed accordingly

import spacy
import re
# Import the Matcher
from spacy.matcher import Matcher
nlp = spacy.load("en_core_web_sm")
#doc = nlp(data)

reviews = list(df.reviews)
reviews

para_data = []
tokenized_data = []
for i in reviews:
  tokenized = sent_tokenize(i)
  #tokenized = sent_tokenize(re.sub(r'[^\w\s]', '', i.lower()) )
  tokenized_data.append(tokenized)
  para_data.append(re.sub(r'[^\w\s]', '', i.lower()) )
```

```python
tokenized_data[1]
```

```python
## Function to convert list elements into string
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1
```

```python
def listToString1(s):  
    
    # initialize an empty string 
    str1 = " , "  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1
```

```python
review_list = []

for i in tokenized_data:
  review_list.append(listToString(i))
```

```python
review_list
```

```python
## Adjective-noun Pattern based matching extraction
adj_noun_list  = []

for i in review_list:
  doc = nlp(i)
  matcher = Matcher(nlp.vocab)
  pattern = [{"POS": "ADJ"}, {"POS": "NOUN"}]
  matcher.add("id1", None, pattern)
  matches = matcher(doc)
  # We will show you the first 20 matches
  print("Matches:", list([doc[start:end].text for match_id, start, end in matches][:20]))  
  adj_noun_list.append(list([doc[start:end].text for match_id, start, end in matches][:20]))
```

```python
# Note: the polarity making data process was done before we performed this extraction was completed. 
# The code to make the polarity making dataset is complex. So we now are making the extracted data compatible 
# to the form the polarity data making code understands. 
```

```python
adj_noun_list
```

```python
## Adverb-adjective-noun Pattern based matching extraction
adv_adj_noun_list  = []

for i in review_list:
  doc = nlp(i)
  matcher = Matcher(nlp.vocab)
  pattern = [{"POS": "ADV"}, {"POS": "ADJ"}, {"POS": "NOUN"}]
  matcher.add("id1", None, pattern)
  matches = matcher(doc)
  # We will show you the first 20 matches
  print("Matches:", list([doc[start:end].text for match_id, start, end in matches][:20]))  
  #adv_adj_noun_list.append(list([doc[start:end].text for match_id, start, end in matches][:20]))
  adv_adj_noun_list.append([doc[start:end].text for match_id, start, end in matches][:20])
```

```python
list_data_adv_adj = []
x = ""
for i in adv_adj_noun_list:
  if len(i) != 0:
    x = "['" + "' , '".join(i) + "']"
    #print(x)
    list_data_adv_adj.append(x)
  else:
    x = ""
    #print(x)
    list_data_adv_adj.append(x)
```

```python
df_adv_adj_noun = pd.DataFrame(list_data_adv_adj)
```

```python
df_adv_adj_noun.to_excel("ADV_ADJ_NOUN.xlsx")
```

```python
list_data_adj = []
x = ""
for i in adj_noun_list:
  if len(i) != 0:
    x = "['" + "' , '".join(i) + "']"
    #print(x)
    list_data_adj.append(x)
  else:
    x = ""
    #print(x)
    list_data_adj.append(x)
```

```python
df_adj_noun = pd.DataFrame(list_data_adj)
```
